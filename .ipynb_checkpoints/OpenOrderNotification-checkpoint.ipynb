# Data Filtering:

Some of the filters applied to the report:
1. indirect lines
2. Lines not in OOR from EMS BI
3. Lines SECE with presidio


```python
def indirect(df):
    if df['Account Category'] =='K' or df['Account Category'] =='P':
        if len(str(df['Shopping Cart Numer']))>3 and (df['Purch. Group']=='301' or df['Purch. Group']=='B03' or df['Purch. Group']=='B02'):
            return 'Indirect'
        else:
            return 'Direct'
    else:
        return 'Direct'

df['Category']=df.apply(indirect,axis=1)
#Excluded Indirect Lines report:
excluded=df.loc[df['Category']=='Indirect']
df=df.loc[df['Category']!='Indirect']



#Creating the key of PO line
df['Line Number']=pd.to_numeric(df['Line Number'], errors='coerce')
df['PO-Line']= df['PO Number']+"-" +df['Line Number'].astype(str)

#Merging with the OOR information
df=pd.merge( left=df, right=oor, how='left',left_on='PO-Line', right_on='PO Line Number')

df['PO Line Number'] = df['PO Line Number'].fillna('notscope')
df = df[df['PO Line Number'] != 'notscope']
df = df.drop_duplicates(subset='PO-Line')

df['OA Vendor'] = df['OA Vendor'].astype(str)
contact['OA Vendor'] = contact['OA Vendor'].astype(str)
#Merging with the Contact list Info
df=pd.merge( left=df, right=contact, how='left',on='OA Vendor')
df['Email'] = df['Email'].fillna('notscope')
df = df[df['Email'] != 'notscope']

```

```python

def presidio(df):
    if str(df['Vendor Number']) =='1000006317' and str(df['SLB Center']) =='SECE':
        return 'Presidio filter'
    else:
        return ''
    
df['pres']=df.apply(presidio,axis=1)
df=df[df['pres']!='Presidio filter']
```

# Prioritization
1. First we create the columns needed for the prioritize:

     a. Past due and Past due Aging
     
     b. Acknowledgment and Aknowledgment aging
     
     c. AMP
     
     d. Domestic-International
     
     

```python
df['AMS']=df['AMS'].map(str)
def OH(df):
    if "OHSUP" in df['AMS']:
        return "Scope"
    elif "|OH" in df['AMS']:
        return "OH"
    else:
        return "Scope"
        
df['OHScope']=df.apply(OH, axis=1)
df = df[df['OHScope'] == 'Scope']
```

```python
def internalFilter(df):
    if "P" in df['OA Vendor'] or "I" in df['OA Vendor']:
        return "Yes"
    else:
        return df['Acknowledged (Yes/No)']
        
df['Final ack']=df.apply(internalFilter, axis=1)

df=df.drop(columns={'Acknowledged (Yes/No)'})
df=df.rename(columns={'Final ack':'Acknowledged (Yes/No)'})
```

```python

#Past Due
df["Delivery/Dock Date (Per Incoterms)"]=df["Delivery/Dock Date (Per Incoterms)"].str.split(" ",n=1,expand=True)[0]
df["DDD"] = pd.to_datetime(df["Delivery/Dock Date (Per Incoterms)"],
                                    format="%Y/%m/%d")



today = pd.datetime.now()
df["Aging Past Due"] = (today - df["DDD"]).dt.days



df["Order Date"]=df["Order Date"].str.split(" ",n=1,expand=True)[0]
#ack
df["PO Order Date"] = pd.to_datetime(df["Order Date"],
                                    format="%Y/%m/%d")


df['Ack Aging']= (today-df['PO Order Date']).dt.days
df.loc[df["Ack Aging"]<0, "Ack Aging"]=0





```

```python
#Remove the CNT AMP
for i, row in df.iterrows():
    if "CNT" in str(row['AMP']) or "NOACT" in str(row['AMP']) :
        df.at[i,'AMP']=""

#AMP Review
def request(df):
    #First review if the AMP was already been touched
    req= ''
    request=''
    if len(str(df['AMP']))>=21:
        if 'AWRVEN' in str(df['AMP'])[21:] or 'WIPVEN' in str(df['AMP'])[21:] :
            request=str(df['AMP'])[str(df['AMP']).find(':')+1:str(df['AMP']).find('-')]
            if request in {'DFP','SKO','SOS', 'STO', 'EXP', 'DEX', 'MRP','NOACT'}:
                req=request
            else:
                req= ''
        else:
            req='actioned'
    elif len(str(df['AMP']))>=5 and  len(str(df['AMP']))<21 and ':' in str(df['AMP']) and '-' in str(df['AMP']):
        request=str(df['AMP'])[str(df['AMP']).find(':')+1:str(df['AMP']).find('-')]
        if request in {'DFP','SKO','SOS', 'STO', 'EXP', 'DEX', 'MRP','NOACT'}:
            req=request
    else:
        req= ''
        
    return req

df['AMP Request']=df.apply(request,axis=1)





# Requested Delivery Date:

def reqDD(df):
    ddd=None
    if df['AMP Request'] in {'DFP','SKO','SOS', 'STO', 'EXP', 'DEX', 'MRP','PRO'}:
        ddnum=df['AMP'][str(df['AMP']).find('-')+1: str(df['AMP']).find('-')+7]
        try:
            ddd= datetime.strptime(ddnum, "%m%d%y").strftime('%m/%d/%Y')
        except:
            ddd=None
    return ddd

        
df['New Requested Delivery Date']= df.apply(reqDD,axis=1)
df['New Requested Delivery Date']=pd.to_datetime(df['New Requested Delivery Date'])
```

```python
def exceptionMessage(df):
    em=''
    if df['AMP Request'] in  {'DFP','SKO','SOS','STO','EXP'} :
        em= 'Expedite'
    elif df['AMP Request'] == 'DEX':
        em= 'De-expedite'
    elif df['AMP Request'] == 'MRP':
        if df['New Requested Delivery Date'] < df["DDD"]:
            em= 'Expedite'
        elif df['New Requested Delivery Date'] > df["DDD"]:
            em= 'De-expedite'
    return em
            

df['Exception Message Flag']= df.apply(exceptionMessage,axis=1)
```

```python
def prioritization(df):
    #Lines without ack, past dues and with AMP of level 1
    if df['Acknowledged (Yes/No)']== 'No' and df['Aging Past Due']>0  and (df['AMP Request']=="DFP" or df['AMP Request']=="SKO" or df['AMP Request']=="SOS" or df['AMP Request']=="STO"):
        return 'Priority 1'
    #Lines not ack for 14 days and past due greater than 14
    elif df['Acknowledged (Yes/No)']=='No' and df["Ack Aging"]>=14 and df["Aging Past Due"]>=14:
        return 'Priority 1'
    
    elif (df['Acknowledged (Yes/No)']== 'No' and df["Ack Aging"]>=14) or (df["Aging Past Due"]>=14):
        return 'Priority 1'
    
    elif (df['AMP Request']=="DFP" or df['AMP Request']=="SKO" or df['AMP Request']=="SOS" or df['AMP Request']=="STO"):
        return 'Priority 1'
    
    elif df['Acknowledged (Yes/No)']== 'No' and df['Aging Past Due']>0  and (df['AMP Request']=="EXP" or df['AMP Request']=="DEX" ):
        return 'Priority 2'

    elif df['Aging Past Due']>0 and (df['AMP Request']=="EXP" or df['AMP Request']=="DEX"):
        return 'Priority 2'
    
    elif df['Acknowledged (Yes/No)']== 'No' and (df['AMP Request']=="EXP" or df['AMP Request']=="DEX"):
        return 'Priority 2'
    
    elif df['AMP Request']=="EXP" or df['AMP Request']=="DEX":
        return 'Priority 2'
    
    elif df['Acknowledged (Yes/No)']== 'No':
        return 'Priority 3'
    
    elif df['Aging Past Due']>0 :
        return 'Priority 3'
    
    elif (df['AMP Request']=="MRP" or df['AMP Request']=="NOACT"):
        return 'Priority 3'
    
    elif df['Aging Past Due']>-30 and df['Vendor Country']=='US':
        return 'Priority 3'
    
    elif df['Aging Past Due']>-90 and df['Vendor Country']!='US':
        return 'Priority 3'
    
    else:
        return 'No priority'
    

df['Priority']=df.apply(prioritization, axis=1)
df = df[df['Priority'] != 'No priority']
```
